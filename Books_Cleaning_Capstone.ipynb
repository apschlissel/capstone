```python
##Import Libraries
import numpy as np
import pandas as pd
```

```python
## Read in Data
original_book_df = pd.read_csv('./books_expanded.csv', sep=',')

original_book_df.shape[0]
```

```python
original_book_df.head()
```

FILTER THE DATA

```python
## Remove Blank Descriptions
filt_df = original_book_df.dropna(subset=['description'])
filt_df.shape[0]
```

```python
## Genre Filters - Definitely
filt_df = filt_df[filt_df['Nonfiction'] != "Nonfiction"]
filt_df = filt_df[filt_df['Erotica'] != "Erotica"]
filt_df.shape[0]
```

```python
## Additional Genre's - Added 6/29
filt_df = filt_df[~filt_df["genres"].str.contains('Parenting')]
filt_df = filt_df[~filt_df["genres"].str.contains('Picture Books')]
filt_df.shape[0]
```

```python
## Genre Filters - Possibly
filt_df = filt_df[filt_df['Biography'] != "Biography"]
filt_df = filt_df[filt_df['Poetry'] != "Poetry"]
filt_df = filt_df[filt_df['genres'] != "[]"]
filt_df.shape[0]
```

```python
## Language Filters
filt_df = filt_df[(filt_df['language'] == "English") | (filt_df['language'] == "")]
filt_df.shape[0]
```

```python
## Book Format Filters
filt_df = filt_df[filt_df['bookFormat'] != "Album"] 
filt_df = filt_df[filt_df['bookFormat'] != "Wattpad"]
filt_df = filt_df[(filt_df['bookFormat'] != "Newsprint")]
filt_df = filt_df[filt_df['bookFormat'] != "Newsletter Serial"]
filt_df = filt_df[(filt_df['bookFormat'] != "Board Book") & (filt_df['bookFormat'] != "Board")]
filt_df.shape[0]
```

```python
filt_df.tail()
```

```python
##Write out the Filtered Dataframe to a .csv
filt_df.to_csv("./books_filtered.csv", sep=',', encoding='utf-8')
```

Reducing to Columns to Pertinent Only

```python
##Removing Columns
reduced_df = filt_df.drop(columns=['rating', 'characters', 'awards', 'numRatings', 'ratingsByStars', 'likedPercent', 'setting', 'coverImg', 'price', 'bbeVotes', 'bbeScore'])
reduced_df = reduced_df.drop(columns=['bookFormat', 'edition', 'publisher', 'firstPublishDate', 'language'])
reduced_df = reduced_df.drop(columns=['Poetry', 'Biography', 'Erotica', 'Nonfiction'])
reduced_df.shape[0]
```
