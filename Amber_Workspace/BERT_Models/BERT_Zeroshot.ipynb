```python
import tensorflow as tf
```

```python
from transformers import pipeline
```

```python
classifier = pipeline("zero-shot-classification")
# classifier = pipeline("zero-shot-classification", device=0) # to utilize GPU
```

```python
candidate_labels = ["university", "relationships", "break ups", "divorce", "weddings", "death", "family", "friendship"]
hypothesis_template = "This book is about {}."
```

```python
import pandas as pd
df = pd.read_csv('text_data_AR.csv')
```

```python
# csv file
df_books_filtered = pd.read_csv('s3://ec2-jupyter-notebook-us-west-2-8c94c42abbd5478ca9a1a477613965a7/books_filtered.csv')
books_filtered_colsreduced = pd.read_csv('s3://ec2-jupyter-notebook-us-west-2-8c94c42abbd5478ca9a1a477613965a7/books_filtered_colsreduced.csv')
```

```python
df = df_books_filtered.sample(frac=1).reset_index(drop=True)
```

```python
ZS_test_set = df[0:10000]
```

```python
df.to_csv('shuffled_df.csv')
```

```python
results = []
counter = 1
for seq in ZS_test_set['description']:
    sequence = seq
    p = classifier(sequence, candidate_labels, multi_label=True)
    print(counter, " sequences complete")
    results.append(p)
    if counter % 1000 == 0:
        r = pd.DataFrame(results)
        r.to_csv('results_{}.csv'.format(counter))
    counter = counter + 1 
ZS_test_set['zero-shot raw'] = results
```

```python
results = pd.DataFrame(results)
results.to_csv('results_1000.csv')
```
